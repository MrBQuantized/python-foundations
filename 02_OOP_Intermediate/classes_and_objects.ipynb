# Object Oriented programming (OOP) — Classes and Objects
---

- 'OOP' is a software design paradigm that organizes code around "objects" rather than logic or functions. It is used to make programs modular(customizable), reusable(), and easier to scale. 
- In simple terms, **OOP**  models real-world entities (such as cars, students, or bank accounts) as objects with their own properties and behaviors


## ⚙️ The Core Concepts of OOP
OOP is built on two main structural components and four foundational principles, often referred to as "pillars".

## 🏛️ Classes and Objects
- **Class** - This is a blueprint or template that defines the properties (data) and behaviors of a class. 
- **Object** - a particular individual instance of a class.(Example: _Dog_ is the class, but the specific _Dog_ is the object)

## 🧱 The 4 Pillars of OOP
- **Encapsulation** — involves the bundling of data and the methods that operate on that data within a single unit (class). There's a restriction to access placed on some of the object's components to protect the data . Direct access to internal data is restricted
- **Abstraction** — Complex and underlying implementation details are hidden, while the important functionalities and necessary part are exposed to the user.
- **Inheritance** — a mechanism that enables one class to acquire the properties and behavior of another class, and therefore,  producing code reusability. New classes are created from existing classes to reuse and extend code.
- **Polymorphism** — The ability of different methods to respond to the same method call in unique, specific ways. The same interface or method name is allowed to behave differently depending on the object.

## Creating Classes & objects

A class is created using the class keyword followed by the name of the class as shown below

Objects are created using a constructor — a callable dunder method for created objects, `__init__`.

```python
Class name: 
def __init__
```

This method takes at least one parameter/argument (must have the '*self*' parameter) which must be included in any function that we define for our class. It represents the object that we want to create. The self parameter is used to access any attribute of the object.

````python
# Create the class
class Person:

    # Call the constructor
    def __init__(self):
        print("Hello World!")

# Create an instance of the class
x = Person()        
````

Let's create attributes

````python
class Person:

    def __init__(self):
        self.name = "Blessing"
        self.age = 25
        
person1 = Person()
print(person1.name)
print(person1.age)
````

We do not want to have the same values for name and age, so we pass more parameters and assign these parameters to the 'self' attributes.

````python
class Person:

    def __init__(self, name, age, height):
        self.name = name
        self.age = age
        self.height = height
       
# Now we call the class and add the required number of arguments in it
person1 = Person("Chika", 32, 190)
print(person1.name)
print(person1.age)
print(person1.height)
````

We can also have another variable 'person2' with different attribute

````python
person2 = Person("Excel", 20, 185)
print(person2.name)
print(person2.age)
print(person2.height)
````

We can also change the values of the attributes


````python
person1.name = 'Henry'
person1.name
````
