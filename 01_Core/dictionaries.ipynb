# Dictionaries
---
## Notes
- Dictionaries allows us to work with `key-value`pairs
- These are two linked values where:
    - `key`: is the unique identifier where we can find out data and 
    - `value`: is that data'
    - A colon `(:)` is used to separate keys from their values

- Just like in a real physical dictionary where you look up the meaning of words, each word represents a key and the definition is the value of the key.
- The key might be a number or a string but the values could be either a string, number or even a list or numbers or strings.

The syntax for dictionaries is shown below:

```python
dict = {'key': 'value'}

# Example
student = {
    'Name':'Onoh Blesing',
    'Age': 25,
    'Nationality': 'Nigerian'
} 
```

````python
# Now lets look at the representation of a student using a dictionary
student = {'name': 'John', 'age': 25, 'courses': ['Maths', 'CompSci']}

print(student)              # This prints out all our keys and values. 

print(student['courses'])   # This prints the values of the key 'courses'
````

````python
# We can also use integers as keys 
students = {
    1: "John",
    2: "Julian",
    3: "Michael"
}

print(students)    
print(students[1])
````

### Accessing a key  and its values 

````python
# Accessing a key that doesn't exist using the get() function.
student = {
    'name': 'John', 
    'age': 25, 
    'courses': ['Maths', 'CompSci']
}

print(student.get('phone'))    # By default this returns 'None' instead of an error
````

### Updating key-value pairs

````python
# We can make custom default values to be returned by passing it as a second argument of get() function.
print(student.get('phone', 'The Key does not exist'))
````

````python
# Adding a new entry to our dictionary

student = {
    'name': 'John', 
    'age': 25, 
    'courses': ['Maths', 'CompSci']
}
student['phone'] = '555-5555'       # This adds a new key-value pair {'phone':555-5555}
student['name'] = 'Jane'            # Updates the value of the key 'name' from 'John' to 'Jane'

print(student)
````

````python
# We can update  multiple keys at once using the update() method 
student.update({
    'name': 'Jane',
    'age': 26,
    'phone': '555-5555'
})
print(student)
````

### Deleting key-value pairs

````python
# Deleting a key and it's value
# This done using the del function

del student['age']
print(student)
````

````python
# We can also use the pop method.
# This method not only deletes the value, but returns it
student.update({
    'name': 'Jane',
    'age': 26,
    'phone': '555-5555'
})

age = student.pop('age')
print(student)
print(age)
````

### length of dictionary

````python
# Using the len() function, we can check the length of the dictionary
student = {
    'name': 'John', 
    'age': 25, 
    'courses': ['Maths', 'CompSci']
}

print(len(student))
````

### Viewing keys or values or both (items)

````python
student = {
    'name': 'John', 
    'age': 25, 
    'courses': ['Maths', 'CompSci']
}

print(student.keys())
print(student.values())
print(student.items())
````

### Looping through keys and values 

````python
student = {
    'name': 'John', 
    'age': 25, 
    'courses': ['Maths', 'CompSci']
}

for key, value in student.items():
    print(key, value)
````
