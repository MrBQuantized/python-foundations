## OS Module
---

The OS module allows us to interact with the underlying operating system in several ways. For instance, we can navigate file system, get file inforamtion, look up and change the environment vairables move files around etc. 

To use this module we must first import using the command

```python
import os
```

````python
import os

print(dir(os))    # dir function shows us all the methods and attributes associated with a module.
````

### Working with directories
1. `.getcwd`: This displays the current working directory
2. `.chdir()`: This changes the directory/path to a new one

````python
print("Get current working directory:", os.getcwd()) 
os.chdir(r'C:\Users\Dell\Documents\MY_WORKSPACE\01_SKILLS\04_PYTHON_MASTERY\01_CORE_AND_OOP')
print("Get the current working directory after changing directory:", os.getcwd())
````

### Inspecting a Directory — listdir() function

The `.listdir()` function is used for displaying all files and folders within a directory.

````python
os.listdir()
````

### Creating a New Folder
There are two methods for doing this
- `os.mkdir()`— Creates just a new main folder only
- `os.makedirs()` — Creates all the intermediate level diretories/folders with in a directory a few levels deep.

We are going to see an illustration below:

````python
print(os.mkdir('OS-Demo-2/Sub-Dir-1'))
#print(os.makedirs('OS-Demo-2/Sub-Dir-1'))
````

````python
print(os.makedirs('OS-Demo-2/Sub-Dir-1'))

print(os.listdir())
````

### Deleting folders 
This is similar to methods of creating a directory.
- `.rmdir`: This deletes/ removes the exact directory.
- `.removedirs`: This deletes/removes all the intermediate directories

⚠️ **Note**: _Deleting directories recursively is more dangerous than creating them. It advisable ti use `rmdir()` when removing dirs so as to delete the exact directory_ 


````python
print(os.removedirs('OS-Demo-2/Sub-Dir-1'))

print(os.listdir())
````

### Renaming files or Folders

In order to rename files or folders, we use the `.rename()`method. The original file name is placed as the first argument followed by the new file name. The syntax for this is:

```python
os.rename('original_filename', 'new_filename')
```

Let's rename the file `text_file.txt` to `text.txt`.

````python
os.rename('test_file.txt', 'text.txt')

print(os.listdir())
````

### Checking File info
`.stat()` reviews all the information about any file.

````python
os.stat('text.txt')
````

We can also use some attributes listed in the output like modification time,`.st_mtime`, and size,  `.st_size` 

````python
print("Modification time:", os.stat('text.txt').st_mtime)
print("size:", os.stat('text.txt').st_size)
````

To print the modification time in datetime format we first import the `datetime` library:

```python
from datatime import datetime 
```

This prints out a human-readable form of the modification time. 

````python
from datetime import datetime as dt   # Alias used to reduce  

# Let's make a variable
mod_time = os.stat('text.txt').st_mtime

print(dt.fromtimestamp(mod_time))   
````

### Displaying the entire directory tree

We can tranverse the entire directory tree and print all the directores and files within using `.
walk()` method. 

```python
os.walk('path')
```
This generator yields a tuple of 3 values as it's walking the directory tree.
For each directory is sees, it yields:
- The directory path
- The directories within that path
- The files within that path

````python
os.chdir(r"c:\Users\Dell\Documents")

# Let's use a simple for-loop

for dirpath, dirnames, filenames in os.walk(r"c:\Users\Dell\Documents\MY_WORKSPACE"):
    print("Current Path:", dirpath)
    print("Directorires:", dirnames)
    print("Files:", filenames)
    print()
````

This is an extremely helful method that can help you keep track of all of the file information within a certain directory structure in a web application.

### Accessing Environment variables

 Assuming you want to acces your home directory location, you can do this by taking the `HOME ENVIRONMENT` variable. We can get environment variables by accessing `os.envrion.get('env_name')`. So, for the HOME environment it becomes:

 ```python
 os.environ.get('HOME')
 # OR USERPROFILE or HOMEPATH on Windows OS
os.environ.get('USERPROFILE')
os.environ.get('HOMEPATH')
 ```

````python
os.chdir(r"c:\Users\Dell\Documents")

print(os.environ.get('USERPROFILE'))
print(os.environ.get('HOMEPATH'))
````

We can use the path returned by the code to create a new file within the HOME (USERPROFILE). One way of doing this is by concatenaion — combining the path and filename;

```python
file_path = os.environ.get('USERPROFILE') + 'test.txt'
```

````python
os.chdir(r"c:\Users\Dell\Documents")

file_path = os.environ.get('USERPROFILE') + 'test.txt'
file_path
````

However, there are drawbacks associated with this method; it's hard to know if all of the forward slashes are in the correct positions or if they are complete. Just as we can see in the output above — double slashes before each directory sub level, and a missing slash before the file name.

In other to prevent this guess work, we can use the `os.path()` module. This module has several useful methods for working with paths, but the method we will be using in this situation is the `os.path.join()`. This method takes in two arguments — the path first, then the text:


````python
file_path = os.path.join(os.environ.get("USERPROFILE"), 'test.txt')

file_path
````

We can see a full path of the file and the slash in the correct location. So, this method removes the guess work out of creating the part and whether you need to add a slash or not.
This method ensures that files ae being read or written to locations properly.

Let's look at other useful methods using the os.path module. For instance, we can get the basename or file name or both from a path (not necessarily a real path)

````python
print("File name:", os.path.basename('/tmp/test.txt'))                   # returns only the file name
print("Directory name:",os.path.dirname('/tmp/test.txt'))                # returns only the directory
print("Directory & File name:", os.path.split('/tmp/test.txt'))          # returns both directory and file name
````

We can also validate the existence of a file and to check if the path is a directory or a file.

````python
print(os.path.exists('/tmp/fgdfgdf'))
print(os.path.isdir('/tmp/fgdfgdf'))
print(os.path.isfile('/tmp/fgdfgdf'))
````

We can also split the file root and the extension using the `.splitext()` method, and the check all the attributes of the os.path module using the `dir()` function

````python
print("Split root and file extension:", os.path.splitext('/tmp/test.txt'))

print(dir(os.path))
````
