## Exploring Modules & The Standard Library
---

### Module:
A module is an independent, reusable chunk of code that performs a specific function. Using modules helps developers organize large programs and makes code easier to maintain.

### Standard Library:
A standard library is a pre-written colection of code, functions, classes, and modules that comes bundled with a programming language. 

**Examples**: `numpy`, `pandas`, `math`, `os`, `json`etc.

We are going to learn how to import modules, starting  by importing modules that we'e created and then, we'll explore a bit of the Standard Library how we can imort those modules to solve problems in python.


Let's look at the module we just created from the file `my_module`:

```python
print('Imported by my_module...')

test = 'Test String'

def find_index(to_search, target):
    '''Find the index of a value in a sequence'''
    for i, value in enumerate(to_search):
        if value == target:
            return i
        
    return -1  
```

Within this module, we can see:
- A print statement which is run together with other codes in a module being imported in another script.
- A variable named '`text`' set to '`Test String`'.
- A function  with two arguments:
    - to_search which is a list and,
    - a target we are looking for.
- A docstring statement telling us the use of the function — the function finds the index of a value in sequence and returns `1` if the index is found. Otherwise, it returns `-1`.


This module can be used in other modules or scripts by importing it first:


````python
import my_module

courses = ['History', 'Math', 'Physics', 'CompSci']
````

Let's try and use the find_module function:

````python
index = my_module.find_index(courses, 'Math')
print(index)
````

### Aliasing 
We can actually specify a name that we can use instead of the full `my_module` name to keep things shorter.

````python
import my_module as mym         

courses = ['History', 'Math', 'Physics', 'CompSci']

index = mym.find_index(courses, 'Math')
index
````

We can import specific functions from a module and use them directory (without the module_name), but this gives us access to only that particular function. We won't be able to use other functions.
```python
from my module import find_index
```
However, other functions can be included by adding commas `,` before the next function:
```python
from my module import find_index, test
```

````python
# Using only the find_index function
from my_module import find_index

index = find_index(courses, 'CompSci')
index
````

````python
# Using both functions
from my_module import find_index, test

index = find_index(courses, 'CompSci')
print(index)
print(test)
````

When importing specific functions from a module, we can also use aliasing, but we must ensure that the aliases are readable and makes sense to other users.

For instance, in the example below, the alias isn't really readable:

````python
from my_module import find_index as fi

index = fi(courses, 'CompSci')          # fi is somehow vague.
index
````

There is a way to import everything in a module. However this method is not recommended because it introduces difficulty in tracking down problems. Anyway, this is done using an asterisk `*` 

```python                       
from my_module import *    
```
Interpreted as *`from my_module, import everything`*


### Locating module's root directory
When a module is imported no path is given. Python finds the location by tranversing multiple locations within a list called `sys.path`. This path can be seen by importing the `sys` module. 

````python
import sys

print(sys.path)
````

The output is a list of directories on your computer/machine where pythn looks for modules when an import is run.

The directory added to the sys.path are added in the order:
1. The directory containing the script being run.
2. Directories listed in the python path environment variable
3. The Standard Library directories
4. The site-packages directory for 3rd-party packages.

It's possible to import the module even if it is in a different location from our sript. One method is by appending it with sys.path()

````python
import sys

sys.path.append(r"/Users/Dell/Desktop/My-Modules")

print(sys.path)
````

````python
from my_module import find_index, test

index = find_index(courses, 'CompSci')
print(index)
print(test)
````

This isn't the best approach, because we are appending this directory before our other imports. In addition if we import our module and you have this hard-coded in multiple locations, then we have to change of those.

- Add a new user variable:
- Name: PYTHONPATH
- Value: C:\Projects\My-Modules
- Restart VS Code / terminal

After the directories in the PYTHONPATH environment variable that we just looked at, sys.path looks at the Standard library directories. This is what makes it possible for us to import modules directly from the standard library. So, when something is part of the Standard Library, it means that we are able to use it without installing anything separately.

The standard library is very important because, it allows us access to various functionalities (written in scripts) of some tasks we want to carry out. We can also be assured that these modules had been written by the best programmers and has been optimized for the best performance.
This doesn't prevent us from trying to write our own versions of these modules for practice, but for production purposes, it's recommended that we use the `tried` and `true` standard library.
