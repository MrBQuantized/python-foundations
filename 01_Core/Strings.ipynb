# Strings - Working with Textual data


Textual data are known as strings. These are variables represented in characters wrapped in quotes (sinlge, double, triple)
For example `'Hello World'` is a string

Variable name must be readable and follow the format:
- `name1`, `name_1`
- `old_name`, `num_of_people`
- `DataFrame` or `bigName`

The following are either not pythonic or invalid:
- `2num`
- `2_people`
- `Number 1`

```python
message = 'Hello World' #  The message holds our text variable which is a string
print(message)
```

```python
# Quotes
message1 = 'Emeka\'s journey'  # we can escape a error arising from single quotes by using a backslash
message2 = "Emeka's journey" # or by using double quotes instead.

message1, message2 
```

```python
# We can also use triple quotes and print our message over multiple lines
message3 = """The fate of the world lies on the shoulders of a young
by from a small village in England
"""
print(message3)
```

```python
# WE can view strings as a string of individual characters:
# symbols — @$%%
# punctuation — space ' ', . ? 
# numerical — 1,2,3.0
# alphabets — a,b,c, d 
# All wrapped in quotes (single, double or triple)

print(len(message))           # returns the length of the string
print(message[:3])            # returns characters starting from index 0 to index 2
```

```python
print(message[:7:2])   # returns first index from 0 up to the 6th and in steps of 2
print(message[:7:-2])
```

```python
print(message[1]) # you can also access individual index
```

```python
# Methods and functions;
# Methods and functions are similar;
# A method is a function belonging to an object

# Case — Upper/lower
print(message.lower())
print(message.upper())
```

```python
# count requires an argument. 
# It returns the number of times a single character or word appear in a string
print(message.count('Hello'))       # 'Hello appears once 
print(message.count('l'))           # 'l' appears 3 times
```

```python
# Find; this tells us the index of a character
print(message.find('e'))            # returns the index 1
print(message.find('World'))        # returns 6 because 'World' begins at index 6
print(message.find('Galaxy'))       # returns -1 for missing characters. 
```

```python
# Replacing characters
# the first argument is/are the character(s) for replacement 
# the second argument is/are the character(s) to be replaced

new_message = message.replace('World', 'Universe')
print(new_message)
print(message.replace('ello', 'i'))
```

```python
# Concatenation

greeting = "Hello"
name = "Michael"

message = greeting +' '+  name    # a string is added between them to space them out
new_message = greeting +', '+ name +'. '+ 'Welcome!'
print(message)
print(new_message)
```

```python
# Formatted string:
# We can make concatenation of strings appear neater using formatted strings
# curly braces are used as placeholders for the variables.

message = '{}, {}. Welcome!'.format(greeting, name)
print(message)
```

```python
# f-string: This doesn't need the .format() clause at the end
# the variables are directly placed into the curly braces
# f is placed in front of the opening quotes 
# which wraps everything (including the curly braces) 

message = f'{greeting}, {name}. Welcome!'
print(message,'\n')
print("Or\n")
print(f'{greeting}, {name}. Welcome!')
```

```python
# This method makes it possible to have codes written inside the 
# curly braces (e.g. {name.upper()})
message = f'{greeting}, {name.upper()}. Welcome!'
print(message)
```

```python
# We can also see an broad overview of all the attributes 
# and methods associated with strings. This is possible using the function
"""dir()
"""
print(dir(name))
```

```python
# We can go further to see what these attributes and methods do by using
# the function
"""help()
"""
# This function works by running it on the class 'string' and not on the variable nae
# It gives us information on what these methods do and the arguments the take.
help(str)
```

```python
# We can also find out more info or description of specific methods.
"""help(str.methodname)
"""
# For example the .lower() method
help(str.lower)
```
