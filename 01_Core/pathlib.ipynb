# Pathlib Module
---
This module is used in place of the `os module`for working with paths. was introduced in python 3.4

## Advantages over os module

- In os module you're working with strings for path which can get messy especially when working between different Operating Systems.

- Pathlib represents file paths as objects rather strings. It will handle those problems better.
- It is more intuitive, readable and less prone to errors.
