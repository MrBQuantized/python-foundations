 ## Functions:
 A function is a set of instructions that are packaged together to perform a specific task

 In order to create a function, we use the `def` keyword followed by the function name and a parentheses where its argument(s) will be placed. An illustration is shown below:

 ```python
 def function_name(arg1,arg2,...):
   # or
 def function_name():   # Without any arguments
 ```


````python
 # The pass keyword can be used if we don't want to make use  of the funcion yet
def hello_func():
    pass           

hello_func()        # parenthesis `()` must be included when calling the function
````

````python
print(hello_func)    # This only prints out the function name and location in memory
print(hello_func())  # The paranthesis () executes the function and returns `None` 
````

````python
def hello_func():
    print('Hello Function!')   

hello_func()
````

## _Functions helps us to reuse codes without repeating ourselves_

Imagine that we printed out some statements 4 times
```python
print('Hello Function!')
print('Hello Function!')
print('Hello Function!')
print('Hello Function!')
```
with and output of : 
```python 
Hello Function!
Hello Function!
Hello Function!
Hello Function!
```
But our boss doesn't want the exclamation marks, but instead a period `(.)`. Instead going through the hard process and editing all the print statements, wecan just update it once in the print statement unser our function and the problem is solved once and for all.

```python
# From
def hello_func():
    print('Hello Function!')

# To
def hello_func():
    return'Hello Function.'
```

This is called keeping your codes _`DRY`_.

## return function
The return function equates to its value. so, instead of printing out the value when thefunction is called, it returns nothing. However, we make use of the `print()` function when we want to execute the function

````python
def hello_func():
    return 'Hello Function.'

print(hello_func()) 
````

````python
# we can treat the returned value as a string using string methods.

def hello_func():
    return 'Hello Function.'

print(hello_func().upper()) 
print(hello_func().lower())
````

````python
# Passing arguments/parameters to our function
# We need to create parameters within our parantheesis
def hello_func(greeting):
    return '{} Function.'.format(greeting)

print(hello_func('Hi'))    # We need to put a statment within the parathesis of hello function
````

````python
# Let's add another parameter `name` with a default value
def hello_func(greeting, name='You'):      # name given a default value
    return '{}, {}'.format(greeting, name)      # another placeholder separated by comma added 

print(hello_func('Hi'))                         # We don't actually need the value of the argument name
````

````python
def hello_func(greeting, name='You'):      
    return '{}, {}'.format(greeting, name) 

print(hello_func('Hi', name='Agnes'))       # We can still write a custom value to name and it would execute instead of the defualt value
````

_Positional arguments should come before keyword arguments_ in the order:

```python
def function_name(*args, **kwargs):
```
- the asterisks `*` and `**` in front of arg and kwargs is for accepting an arbirtrary number of arguments 
- the position argument and keyword argument must not always be written as `args` and `kwarg`. They are conventions that we will see a lot.
- It's recommended to work with conventions so that people can understand your codes.

For instance, let's assume the `student_info()` function takes positional arguments in that represents the classes that the student is taking and the keyword arguments parsed in would be random information about the student.

````python
def student_info(*args, **kwargs): 
    print(args)
    print(kwargs) 

student_info('Maths', 'Art', name='John', age=22)
````

- `args` is a tuple with all of our positional arguments
- `kwargs` is a dictionary with all of our keyword arguments

When you see a function call with arguments using the `*` or `**`, it would upack a sequence or dictionary and parse those values into the funtion individually 

````python
def student_info(*args, **kwargs):
    print(args)
    print(kwargs)

courses = ['Math', 'Art']
info = {'name':'John', 'age': '22'}

student_info(courses, info)
# This passes in the complete list and dictionary as positional arguments in a tuple.
# This is not what we want.
````

````python
def student_info(*args, **kwargs):
    print(args)
    print(kwargs)

courses = ['Math', 'Art']
info = {'name':'John', 'age': '22'}

student_info(*courses, **info)    
# with the asterisks added, the values will be unpacked and pass them individually
````

## Examples:

````python
# Number of days per month. First value placeholder for indexing purposes.
month_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
 
def is_leap(year):
    """Return True for leap years, False for no-leap years."""

    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def days_in_month(year, month):
    """Return number of days in that month in that year."""

    # year 2017
    # month 2
    if not 1 <= month <=12:
        return 'Invalid Month'
    
    if month == 2 and is_leap(year):
        return 29
    
    return month_days[month]

print(is_leap(2017))
````

````python
print(days_in_month(2017, 2))
````
