# Working with Numeric Data: Integers and Floats

- `Integers`: These are whole numbers. They can only have distinct values (e.g. 1, 2, 100, 17)
- `Float`: These are decimals—they can have any value in the number line.

```python
num1 = 5
num2 = 2.0
print(type(num1))
print(type(num2))
```

```python
# Arithmetic Operators:
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)
print("Floor Division:", num1 // num2) # Returns only the quotient (without any remainder)
print("Exponent:", num1 ** num2)
print("Modulus:", num1 % num2)   # Returns only remainder of the quotient 
```

```python
num = 1
num += 1        # incremental operation by 1

print(num)
```

```python
# Absolute value
print(abs(-3))
```

```python
# Round
print(round(3.75))
print(round(3.759, 2)) # Round the number to the 2nd digit after the decimal. 
```

```python
# Comparisons: These comparisons return booleans which are True/False values


num_1 = 3
num_2 = 2
print("Equal?", num_1 == num_2)
print("Not Equal?", num_1 != num_2)
print("Greater than?", num_1 > num_2)
print(":Less Than?", num_1 < num_2)
print("Greater or Equal?", num_1 >= num_2)
print("Less or Equal?", num_1 <= num_2)
```

```python
num_1 = '100'
num_2 = '200'

print(num_1 + num_2) 

# The result are not as expect because the variables are strings
# Addition of strings concatenates them (combines the strings instead of the arithmetic operation of summing)
```

```python
# However we can cast these variables and assign the datatype we want (or wrapping them in the
# datatype int()/float()) and perform arithmetic operations on them

num_1 = int(num_1) 
num_2 = int(num_2)
print(num_1 + num_2)
print(abs(num_1 - num_2))
```
