# Lists, Tuples, & Sets

## Lists:
This allows us to work with a list of values or items.

````python
# lets look at a list of courses
courses = ['History', 'Math', 'Physics', 'CompSci']

print(courses)
````

### Indexing: Accessing items in a list

````python
# To access the location of values in our lists we use square brackets
# index: [0,1,2,3] 
# negative index: [-4,-3,-2,-1]
# position: [1st,2nd,3rd,4th]
# For printing out a range of values/items, we use the format:
"""variable[start:stop:step]
"""
# Where:
# -- start: staring index (included)
# -- stop : stopping index (excluded)
# -- step: number of steps 

print(courses[0])       # prints out the value at index 0 (History)
print(courses[2])       # prints ou the value at index 2 (Physics)
print(courses[-1])      # prints out the last value (Co mpSci)
print(courses[::-1])    # prints out all items in the list starting from the last to the first

# It is usually more convenient to print out the last item in a list, 
# especially if the list is very long.
````

### Adding items to a list

````python
# Append: Adding items to the list
courses.append('Art')
print(courses)
````

````python
# Insert: adding at specific position
courses.insert(0, 'Chemistry')
print(courses)
````

````python
# Extend: we use this method when we want to add
# multiple values to our list
courses = ['History', 'Math', 'Physics', 'CompSci']
courses_2 = ['Art', 'Education']

courses.insert(0, courses_2)   
print(courses)
# this adds the entire list so that we have a list within a list
# instead of the individual items
# We will correct this using extend method 
````

````python
courses = ['History', 'Math', 'Physics', 'CompSci']
courses_2 = ['Art', 'Education']

courses.extend(courses_2)   
print(courses)
````

### Removing items from a list

````python
# Remove: 
# This removes a specific values
courses = ['History', 'Math', 'Physics', 'CompSci']
courses.remove('Math')

courses
````

````python
# Pop:removes the last value by default
courses = ['History', 'Math', 'Physics', 'CompSci']
courses.pop()
print(courses)
````

````python
# It also returns the value/item it removed
courses = ['History', 'Math', 'Physics', 'CompSci']
popped = courses.pop()
print(popped)
print(courses)
````

### Sorting 

````python
courses = ['History', 'Math', 'Physics', 'CompSci']
courses.reverse()
print(courses)
````

````python
# Sort:
# strings in alphabetical order
# numbers in descending order. 
courses = ['History', 'Math', 'Physics', 'CompSci']
# It also sort numbers 
nums = [1,5,2,4,3]
nums.sort()
courses.sort()
print(courses)
print(nums)
````

````python
# We can also reverse our sorted list. An easy way of doing this is to pass 
# an argument 'reverse' in the sorted method
courses = ['History', 'Math', 'Physics', 'CompSci'] 
nums = [1,5,2,4,3]
nums.sort(reverse=True)
courses.sort(reverse=True)
print(courses)
print(nums)
````

````python
# We can also get the reversed version of our list 
# without altering the original list
# We use the sorted() function.
"""sorted(variable)
"""
courses = ['History', 'Math', 'Physics', 'CompSci'] 
nums = [1,5,2,4,3]

sorted_courses = sorted(courses)
sorted_nums = sorted(nums)

print(sorted_courses)
print(sorted_nums)
````

## Arithmetic methods

````python
# Min and Max values 
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 
nums = [1,5,2,4,3]
print("Maximum number:",  max(nums))
print("Minimum number:", min(nums))
print("Sum of numbers:", sum(nums))
````

### Finding index of values

````python
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 

print(courses.index('Physics'))   # returns the index 2
print(courses.index('Art'))       # returns an error because the item 'Art' is not in the list
````

````python
print('Arts' in courses)
# checks if the item is in the list. This would return a boolean 'False'
print('Math' in courses)
````

````python
# We can also this statement to loop through values in our list by using a `for`` loop
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 

for item in courses:
    print(item)             
    # This prints each item in our loop. It accesses each value line by line
````

````python
# We can alo access the value and index of the item in python
# This is done by using the eumerate() function:
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 

for index, course in enumerate(courses):
    print(index, course)
````

````python
# We can also pass in a custom start index in the second argument of enumerate():
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 

for index, course in enumerate(courses, start=1):
    print(index, course)
````

````python
# We can our list into a string of comma-separated-values (CSV);
# THis is done using the 'join' function.
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 
courses_str = ', '.join(courses)            # comma-separated-values
courses_str_hyphened  = ' - '.join(courses) # list items are separated by hyphen

print(courses_str)
print(courses_str_hyphened)
````

````python
# We can also do the reverse version by converting the string back to a list
courses = ['History', 'Math', 'Physics', 'CompSci', 'Education'] 
courses_str = ', '.join(courses)          
courses_str_hyphened = ' - '.join(courses)
new_list = courses_str_hyphened.split(' - ')     # this splits the values hyphened and makes a new list out them

print(courses_str_hyphened)
print(new_list)
````

## Tuples:
- They are similar to lists but they are immutable

````python
list_1 = ['History', 'Math', 'Physics', 'CompSci']
list_2 = list_1

print(list_1)
print(list_2)

list_1[0] = 'Art'

print(list_1)
print(list_2)
````

````python
# Immutable
tuple_1 = ('History', 'Math', 'Physics', 'CompSci')
tuple_2 = tuple_1

print(tuple_1)
print(tuple_2)

tuple_1[0] = 'Art'          # This is impossible to do with a tuple and will result to an error.

print(tuple_1)
print(tuple_2)
````

## Sets
### Info
- Sets contain values that are:
    - unordered
    - cannot be duplicated
    - immutable

- They are used to test if an item is part of test. This is called a _`Membership Test`_.
- They are optimized for membership test; It's more effective than lists and tuple.

````python
cs_courses = {'History', 'Math', 'Physic', 'CompSci'}

print(cs_courses)
````

````python
# Suppose we add a duplicate of an item in the set:
cs_courses = {'History', 'Math', 'Physic', 'CompSci', 'Math'}

print(cs_courses)       # This will still return only distinct values (no duplicates)
````

````python
# Membership test 
cs_courses = {'History', 'Math', 'Physic', 'CompSci', 'Math'}

print('Math'in cs_courses)   
````

### Relationships between sets
- `intersection()`: returns items that are in both  cell:
    ```python
    set_1.intersection(set_2)  # items in et_1 that are also in set_2
    ```
- `difference()`: returns items that are in one set but not in the other. The syntax is:
    ```python
    set_1.difference(set_2)  # items in set_1 but not in set_2 
    ```

- `union()`: combines both sets and prints all the items but removes duplicates:
    ```python
    set_1.union(set_2)  # combines items in set_1 and set_2 
    ```

````python
# We can also compare between sets using function;
art_courses = {'History', 'Math', 'Art', 'Design'}
cs_courses = {'History', 'Math', 'Physic', 'CompSci', 'Math'}

print("Intersection of CS and Art courses:", cs_courses.intersection(art_courses))   
print("Difference of CS and Art courses:", cs_courses.difference(art_courses))
print("Union of CS and Art courses:", cs_courses.union(art_courses))  # Notice that there are no duplicates
````

## Creating Empty Lists, Tuples and Sets

We can create the empty lists, tuples and sets by either assigning empty square brackets or built-in class to a variable. This is illustrated below:
```python 
# Empty List
empty_list = []         # empty square brackets
empty_list = list()     # built-in list class

# Empty Tuples
empty_tuple = ()        # empty brackets
empty_tuple = tuple()   # built-in tuple class

# Empty Set
empty_set = set()       # built_in set class

empty_set = {}      # This creates an empty dictionary instead.
```
