# Conditionals and Booleans — If, Elif and Else statements

- Conditionals  are used to determine what codes or statements get executed depending on whether certain values evaluate to True or False. 
- These True and False values are called _`Booleans`_.


Let's consider a hard-coded condition with True and False obvious conditional statements

```python
# Let's look at an obvious if statement.
if True:
    print("Conditional was True")   # This prints out the message because the statement is True.
```

```python
# Now considering if he condition is False.
if False:
    print("Conditional was True")   # This returns nothing because our condition is obviously False
```

_`Conditional statements are not usually hard-coded as shown above`_.

Let's look at actual conditional statements that evaluate to either True or False.

```python
language = 'Python'

if language == 'Python':
    print('Conditional was True')
```

## Comparators

In Python,comparators (also called comparison operators )are used to compare two values. They return either `True` of `False`

- Equal:                  `==`
- Not Equal:              `!=`
- Greater Than:           `>`
- Less Than:              `<`
- Greater or Equal:       `>=`
- Less or Equal:          `<=`
- Object Identity:        `is`


```python
# Checking a single a language
language = 'Python'

if language == 'Python':
    print('Language is Python')
else:
    print('No match')
```

```python
language = 'Java'

if language == 'Python':
    print('Language is Python')
else:
    print('No match')
```

```python
# Checking for multiple languages
language = 'Java'

if language == 'Python':
    print('Language is Python')
elif language == 'Java':
    print('Language is Java')
elif Language == 'Javascript':
    print('Language is Javascript')
else:
    print('No match')
```

## Logical operators:
These are combined or modify conditions that evaluate to `True` of `False`.

- `and` — Both conditions must be `True`. 
Example: 5 > 3 and 8 < 4 evaluates to `True`

- `or` — At least one condition must be True. 
Example: 5 > 3 or 2 > 4 evaluates to `True`

- `not` — Reverses the truth value. Example not(5 > 3) evaluates to `False`.

```python
# Case 1: 2 condition evaluates to True
user = 'Admin'
logged_in = True

if user == 'Admin' and logged_in:
    print('Admin Page')
else:
    print('Bad Credentials')
```

```python
user = 'Admin'
logged_in = False

if user == 'Admin' and logged_in:
    print('Admin Page')
else:
    print('Bad Credentials')
```

```python
# Case 2: Any one of two condition evaluates to True
user = 'Admin'
logged_in = False

if user == 'Admin' or logged_in:
    print('Admin Page')
else:
    print('Bad Credentials')
```

```python
# Case 3: Switching truth value using `not`
user = 'Admin'
logged_in = False

if not logged_in:               # The not switches the False to True and so the code is executed
    print('Please log in')
else:
    print('Welcome')    
```

```python
a = [1,2,3]
b = [1, 2, 3]

print(a == b)   # This returns True because the values in the list are the same
print(a is b)   # This returns False because a and b are different objects in memory
print(id(a))    # This returns the memory address of a
print(id(b))    # This returns the memory address of b, which is different from a   
```

```python
# If we equate b = a then they will point to the same memory location and so both `a is b` and `a == b` will return True
a = [1,2,3]
b = a

print(id(a))    # This returns the memory address of a
print(id(b))    # This returns the memory address of b, which is the same as
print(a == b)   # This returns True because the values in the list are the same
```

```python
a = [1,2,3]
b = [1, 2, 3]

# (a is b) is the same as id(a) == id(b)
# Let's see the illustration
print(a is b)   # This returns False because a and b are different objects in memory
print(id(a) == id(b))    # This returns False because a and b are different objects in memory
```

### False values:

To determine what Python evaluates to True or False, it's easier to show everything that evaluates to False, then by default, everythin else will evaluate to True. 

- False
- None
- Zero of any numeric type
- Any empty sequence. For example, `''`, `()`, `[]`.
- Any empty mapping. For example, `{}` 

```python
# Now let's see what python evaluates to true or false
condition = False

if condition:
    print('Evaluated to True')
else:
    print('Evaluated to False')
```

```python
# None evaluates to false as well
condition = None

if condition:
    print('Evaluated to True')
else:
    print('Evaluated to False')
```

```python
# Zero numeric value evaluates to False
condition_1 = None

print("output1:")

if condition_1:
    print('Evaluated to True')
else:
    print('Evaluated to False')

print("")

# While other non-zero numeric values evaluate to True

print("output2:")
condition_2 = 10

if condition_2:
    print('Evaluated to True')
else:
    print('Evaluated to False')
```

```python
# Emptyy sequence (lists, tuple, string) and empty mappings evaluate to False.

condition_seq1 = ''
condition_seq2 = []
condition_seq3 = ()
condition_mapping = {}

# Empty string
print("ouptut1:")

if condition_seq1:
    print('Evaluated to True')
else:
    print('Evaluated to False')

print('')

# Empty List
print('output2:')
if condition_seq1:
    print('Evaluated to True')
else:
    print('Evaluated to False')

print('')

# Empty tuple
print('output3:')

if condition_seq1:
    print('Evaluated to True')
else:
    print('Evaluated to False')

print('')


# Empty mapping
print('output4:')

if condition_mapping:
    print('Evaluated to True')
else:
    print('Evaluated to False')
```

If we insert values into the following sequences and mappings then we would have True values

```python
# For instance:
condition_mapping = {'Name': 'Blessing'}

if condition_mapping:
    print('Evaluated to True')
else:
    print('Evaluated to False')
```
