Loops and Iterations — For & While Loops
</h1>

```python
nums = [1,2,3,4,5]

for num in nums:
    print(num)   

# This loops through all the items in our list one by one. 
# Each time the num variable will be equal to the next item.
```

### Important Keywords:

- `Break` — This completely breaks out of the loop
- `continue` — moves on to the next iteration of our loop

```python
# Break statement: Suppose wearelooking for a number (3) and when it's found the loop stops.

nums = [1,2,3,4,5]

for num in nums:
    if num == 3:
        print('Found!')
        break
    print(num)   
```

```python
# Continue Statement: If want to ignore a value and not break out of the loop
nums = [1,2,3,4,5]

for num in nums:
    if num == 3:
        print('Found!')
        continue
    print(num)   
```

### Nested Loop — Loop within a Loop

In the example below, for each number, it will loop through every character in the string and print out the character and the number and then moves on to the next number and repeats the process

```python
nums = [1,2,3,4,5]

for num in nums:
    for letter in 'abc':
        print(num, letter)
```

```python
# Looping for a particular number of times (say 10 times)
# This is done using the range() function

for i in range(10):
    print(i)
```

```python
# We could include starting, stopping and indexes 
for x in range(1, 11):     # 1 to 11 because the stopping index isn't included
    print(x)
```

## While Loops:

Our `for-loops` iterated through a certain number of values but `while-loops` keep repeating until a certain condition is met.

```python
x = 0
while x < 10:          # goes through the loop
    print(x)           # prints the value
    x += 1             # then increases the value by 1 and then repeats the process
 
# This iteration continues until x is no longer less than 10
```

```python
# we can also break out of the while loop
m = 0

while m < 10:
    if m == 5:
        break
    print(m)
    m += 1
```

```python
# Replacing the comparison ( m < 10) with the value of 'True' will lead to 
# an infinite loop which would only be terminated using a break statement
m = 0

while True:
    if m == 5:
        break
    print(m)
    m += 1
```
