## Enum
Enum is a data type which links a name to an index. They are usefrul to represent a closed set of options.

```python
from enum import Enum

class QhBrowserAction(Enum):
    QUERY_BUTTON_CLICKED = 1
    SAVE_BUTTON_CLICKED = 2
    DATE_CHANGED = 3
    QH_NAME_CHANGED = 4
    SLIDER_MOVED = 5

a = QhBrowserAction.DATE_CHANGED
b = QhBrowserAction.QH_NAME_CHANGED
a.name, a.value, b.name, b.value
```

```python
a_next = QhBrowserAction(a.value + 1)
a_next
```

```python
if a_next == QhBrowserAction.QH_NAME_CHANGED:
    print('In state {}'.format(a_next.value))
```

### Copying lists

```python
x = [1, 2, 3, 4]
y = x
print(x, y)
```

```python
x,y
```

```python
x.copy()
```

```python
x = [1, 2, 3, 4]
y = x.copy()       
print(x, y)
```

```python
x = [[1, 'a'], 2, 3, 4]
y = x.copy()        # equivalent to x[:]
y[0] = 'a'
print(x, y)
```

```python
x = [[1, 'a'], 2, 3, 4]
y = x.copy()      
y[0][0] = 'b'
print(x, y)
```

The reason for this behavior is that python performs a shallow copy

```python
from copy import deepcopy
x = [[1, 'a'], 2, 3, 4]
y = deepcopy(x)
y[0][0] = 'b'
print(x, y)
```

### Sorting lists — inplace operations


```python
x = [1, 10, 2, 9, 3, 8, 4, 6, 5]
x = x.sort()
print(x)
```

`list.sort()` is an inplace operation. In general, inplace operations are efficient as they do not create a new copy in memory

## Sets

```python
# Sets are immutable and contain only unique elements
{1,2,3,4}
```

```python
{1,2,3,4,4}   # Will remove the duplicated 4
```

It is a neat way to obtain unique elements in a list

```python
my_list = [1,2,3,4,5,5,5]
set(my_list)
```

```python
# or a tuple
my_tuple = [1,2,3,4,4,5,5,5]
set(my_tuple)
```

You can perform set operations on sets

```python
A = {1,2,3}
B = {3,4,5}
print(f'A + B = {A.union(B)}')
print(f'A | B = {A.intersection(B)}')
print(f'A - B = {A.difference(B)}')
print(f'A ▲ B = {A.symmetric_difference(B)}')
```

```python
# Alternatively
print(f'A + B = {A | B}')
print(f'A | B = {A & B}')
print(f'A - B = {A - B}')
print(f'A ▲ B = {A ^ B}')
```

```python
print(f'A * 0 = {A.intersection({})}')
```
