## File objects


Whether you use python for desktop or web applications, you likely going to be interacting with files. So, It is going to be a good skill learning how to properly interact with these file objects on python.

To get a file object, we can use the built-in open command. The `open()` command allows us to specify whether we want to open the file for:
- reading 
- writing 
- appending or 
- reading and writing.

Theses specifications can be shown in our open command by a second argument.


```python
file = open('file_name' 'r')    # for reading a file
file = open('file_name' 'w')    # for writing to a file
file = open('file_name' 'a')    # for appending to a file
file = open('file_name' 'r+')   # for reading and appending to a file
```
```Plain text
If we do not specify anything then, it opens the file for reading by default.

The way we are going to open this file isn't normally recommended. It's recommended to use a context manager which we would be looking into in a bit.
```

````python
f = open('test_file.txt', 'r')

print(f.mode)       # print the mode with which the file was opened — r, w, a
print(f.name)       # prints out the file's name
f.close()           # closes the file
````

## Opening a file using a context manager

- In using a context manager, we use the `with` keyword together with the `open()` command.
- Context manager allows us to work with the file from within its code block.
- When we exit this block of codes, it would automatically close the file.
- It also closes the file if any exception is thrown.

````python
with open('test_file.txt', 'r') as f:
    pass

print(f.closed)         # This confirms that file is automatically closed when exiting the code block.
````

````python
with open('test_file.txt', 'r') as f:
    f_contents = f.read()
    print(f_contents)           # prints the contents
````

For large files we have other methods:

**Example**: *`readlines()`  prints all lines out as a different element of the file.*

````python
with open('test_file.txt', 'r') as f:
    f_contents = f.readlines()   
    print(f_contents) 
````

_**readline ()** grabs the first line of our file_

````python

with open('test_file.txt', 'r') as f:
    f_contents = f.readline()   
    print(f_contents)  

    f_contents = f.readline()   
    print(f_contents)  
````

Each time you use the readline(), it prints a new empty line before the next line in your file. We can stop the continues printing of a new line by adding a second argument with an empty string, `end=''` in the print command

````python
with open('test_file.txt', 'r') as f:
    f_contents = f.readline()   
    print(f_contents, end='')  

    f_contents = f.readline()  
    print(f_contents, end='') 
````

We have not yet solved the problem of reading all of the contents of a very an extremely large file. If we read the entire file and all at once, we could run out of memory. So, instead of using `readline()` continuously, we can iterate over the lines on a file. This can be solved by using a `for-loop`.

````python
# This would go through and get one line at a time, so we don't have a memory issue.
with open('test_file.txt', 'r') as f:

    for line in f:
        print(line, end='')
````

This is good enough for most people, but sometimes, we might want more control over what exactly we're reading from the file.
We can get the same result by using the f.read() command and we also specify the amount of data we want to be read at a time by passing it as an argument.

E.g:
```python
f_contents = f.read(100)   # The first 100 characters
```

````python
with open('test_file.txt', 'r') as f:

    f_contents = f.read()
    print(f_contents, end='')
````

````python
# the first 100 characters

with open('test_file.txt', 'r') as f:

    f_contents = f.read(100)
    print(f_contents, end='')
````

````python
with open('test_file.txt', 'r') as f:

    f_contents = f.read(100)        
    print(f_contents, end='')
    
    f_contents = f.read(100)        # Adding the second f.read statement prints the rest of the file.
    print(f_contents, end='')

# A third line containing the same come will return nothing.
````

## Reading large files

For very large files, since we do not know exactly how long the file would be, we are going to use a kind of loop that iterates over small chunks at a time.

````python
with open('test_file.txt', 'r') as f:

    size_to_read = 100

    f_contents = f.read(size_to_read)

    while len(f_contents) > 0:
        print(f_contents, end='')
        f_contents = f.read(size_to_read)
 
````

````python
# We can change the character size to 10 and add * to after each 10 characters
# for us to get a better understanding of the loop
with open('test_file.txt', 'r') as f:

    size_to_read = 10

    f_contents = f.read(size_to_read)

    while len(f_contents) > 0:
        print(f_contents, end='*')
        f_contents = f.read(size_to_read)
````

This result shows us that it's printing 10 characters one at a time

````python
# We see the position when reading the file by using .tell() method.
# first 10 characters

with open('test_file.txt', 'r') as f:

    size_to_read = 10

    f_contents = f.read(size_to_read)
    print(f.tell())

# We can manipulate our current position using a .seek() method
````

````python
# First 20 characters
with open('test_file.txt', 'r') as f:

    size_to_read = 10

    f_contents = f.read(size_to_read)
    print(f_contents, end='')

    f_contents = f.read(size_to_read)
    print(f_contents)
````

This printed out first the 20 characters from our file. When we read in the 2nd chunk, it picked up from the 10th position and read the next 10 finals.

We can also make the reading to start right from the beginning using `.seek()` method, but it can be used to change to any starting position of choice.

````python
with open('test_file.txt', 'r') as f:

    size_to_read = 10

    f_contents = f.read(size_to_read)
    print(f_contents, end='')
    
    f.seek(0)       # resets the reading position to start from the beginning
    
    f_contents = f.read(size_to_read)
    print(f_contents)
````

## Writing to file

### Note:
- The file must opened in write mode using th keyword argument `'w'`.
- If the file doesn't exist already, then opening in write mode would automatically create it. However, if it does exists, it would be overwritten.
- To avoid overwriting any file, it should be opened in appending mode with the argument `'a'`.

````python
# Creating a new file
with open('test2_file.txt', 'w') as f:
    pass
````

````python
# Writing to the file
with open('test2_file.txt', 'w') as f:
    f.write("Test")
````

````python
with open('test2_file.txt', 'w') as f:
    f.write("Test")
    f.write("Test")

# This prints Test strings back-to-back.
````

`seek()` can also be used when writing files to set the postion back at the beginning of the file. 

````python
with open('test2_file.txt', 'w') as f:
    f.write("Test")
    f.seek(0)
    f.write("Test")
````

One might run into confusion while using .seek() in writing to files, because it does not overwrite everything, but rather, what it needs to overwrite. For instance another write command with 'R' as the value may replace the 'T' in the first string 'Test' and not the whole string.

````python
with open('test2_file.txt', 'w') as f:
    print('Before changing writing position:',f.write("Test"))
    f.seek(0)               # changes the postion back to the beginning
    print('After position change:',f.write("R"))     # Replaces 'T' with 'R' in 'Test'
   
file = open('test2_file.txt', 'r')
print('New text:',file.read())  
file.close()                
````

### Reading and Writing

Let's put everything together and use read and write on multiple files simultaneouslly. We will make a copy of our `text_file.txt` and delete the `text2_file.txt` so as to avoid confusion between the two.

````python
with open('test_file.txt', 'r') as rf:
    with open('test_copy.txt', 'w') as wf:
        for line in rf:
            wf.write(line)
````

The code above contains a nested codes—the first opening the original file and the second code copying the contents of the original file to the second file which is yet to be created. The loop interprets as _For each line in the original file `(rf)`, write that line the second file `(wf)`.

The result gives us an exact copy of the original file 

### Working with images
Let's look at something similar and try to copy a large picture file. 
- Image files are opened in binary mode — reading and writing bytes instead of strings otherwise, we will run into an exception `UnicodeDEcodeError`.
- To use binary mode, we append `b` to our read `('r')` and write `('w')` so that we have `rb` and `wb` as our arguments

````python
with open('Slides.png', 'rb') as rf:
    with open('Slides_copy.png', 'wb') as wf:
        for line in rf:
            wf.write(line)
````

Instead of doing this line-by-line, we can do it in chunk sizes:


````python
with open('Slides.png', 'rb') as rf:
    with open('Slides_copy.png', 'wb') as wf:
        chunk_size = 4096
        rf_chunk = rf.read(chunk_size)
        while len(rf_chunk) > 0:
            wf.write(rf_chunk)
            rf_chunk = rf.read(chunk_size)
````
