
As a developer you won't want end users to see such error messages so, it is paramount that we use `try` and `except` blocks to handle the sections of our codes we anticipate might throw an error or exception.

Let's try it with this code above

````python
try:
    f = open('testfile.txt')
except Exception:
    print('Sorry. The file does not exist')
````

This returns our custom error instead of the longer error from python.

However, the goal of ***try and except*** is to get specific exceptions and not so that we can get all the exceptions we can run into.

```python
try:
    f = open('testfile.txt')
except Exception:
    print('Sorry. The file does not exist')
```

`Exception` in the code above is not going to catch only the exceptions we anticipate but all of the exceptions and so, we need to make custom exceptions for specific ones.

Let's try and make the filename correct and then, insert a wrong statement below it.

````python
try:
    f = open('test_file.txt')       
    var = bad_var
except Exception:                  
    print('Sorry. The file does not exist') 
````

It still returned the same error message because of the vague and general exception.

````python
try:
    f = open('test_file.txt')
    var = bad_var
except FileNotFoundError:
    print('Sorry. The file does not exist')
````

We can see that file opened correctly and so the `FileNotFoundError` was skipped, but then we caught the name error `bad_var` and it returned the default python traceback

We can also insert another exception below the first one, but we must ensure that when handling multiple exceptions such as this, the more specific exception must be kept at the top and the general ones, the further you go down. 

````python
try:
    f = open('test_file.txt')
    var = bad_var
except FileNotFoundError:
    print('Sorry. The file does not exist')
except Exception:
    print('Sorry. Something went wrong')
````

We can also use an alias to print out the actual exception instead of a custom print message.

````python
try:
    f = open('test_file.txt')
    var = bad_var
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)
````

Let's also treat the `else` and `finallly` block.

## `else` and `finally` blocks

### `else` Clause:
- This runs codes that needs to be executed if the `try` clause does not raise an exception.
- It only runs if we don't throw an exception.

Let's try to open the file and and read it

````python
try:
    f = open('test_file.txt')       # try clause doesn't throw an exception because the filename is correct. 
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)
else:
    print(f.read())     # prints out the content
    f.close()           # closes the file
````

### `finally` Clause 
- Unlike the `else` clause, the `finally` clause runs regardless of what happens—whether there's an exception or the code runs successfully.
- It is important so that we can release certain resources no matter what happens in the try and except blocks. For instance, if you were working on a database, this would be the point where it is being closed.
- In a nutshell, it handles something that needs to be done regardless if the codes were successful or not.

````python
# Executing with no exceptions
try:
    f = open('test_file.txt')
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("Executing Finally...")
````

````python
# Executing with an exception
try:
    f = open('testfile.txt')
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("Executing Finally...")
````

We can also manually raise exceptions. For instance:

````python
# Let's raise an exception from the file bad_file.txt
try:
    f = open('bad_file.txt') 
    if f.name == 'bad_file.txt':        
        raise Exception             # raises the exception
except FileNotFoundError as e:
    print(e)
except Exception as e:
    print('Error!')
else:
    print(f.read())
    f.close()
finally:
    print("Executing Finally...")
````
