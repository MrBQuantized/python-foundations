### Identity operators

Identity operators are used to compare objects. The different types of identity operators are is and is not, which are defined as follows:

````python
# The `is`` operator is used to check whether two variables refer to the same object

list1 = []
list2 = []

if list1 == list2:
    print("Both are equal")
else:
    print("Both are not equal")
if list1 is list2:
    print("Both variables are pointing to the same object")
else:
    print("Both variables are not pointing to the same object")


list3 = list1

if list3 is list2:
    print("Both are pointing to the same object")
else:
    print("Both are not pointing to the same object")
````

````python
# The `is not` operator is checks whether two variables point to the same object or not. Outputs `True` if both do, otherwise, it `False``:

list1 = []
list2 = []

if list1 is not list2:
    print("Both list1 and list2 variables are the same object")
else:
    print("Both list1 and list2 variables are not the same object")
````

### Logical operators

combine conditional statements `(True or False)`. There are three types of logical oprators: `AND`, `OR`, and `NOT`

````python
# for AND ops both statements must be true.
a = 32
b = 132

if a > 0 and b > 0:
    print("Both a and b are greater than zero")
else:
    print("At least one variable is less than 0")
````

````python
# OR returns true if any of the statements are true.
a = 32 
b = -32
if a > 0 or b > 0:
    print("At least on variable is greater tha zero")
else:
    print("Both variables are less than 0")
````

````python
# NOT operator returns True if the object is false, 
# otherwise it returns False

if not a:
    print("Boolean value of a is False")
else:
    print("Boolean value of a is False")
````

### Complex data types
dictionary, set data types: set and frozenset

#### Dictionaries

similar to list in the sense of it being a collection/container of objects, data is unordered, {key-value} paired, hashable, immutable accepts any arbitrary Python objects.


An object is `hashable` if it has a hash values that remains unchanged during its lifetime in the program.

E.g:

| object | hash value |
|--------|------------|
| `12`     |` # int`     |
| `'winter'` | `# str`   |

````python
# Creating a dictionary and accessing elements

person = {}     

print(type(person))

person['name'] = 'ABC'
person['lastname'] = 'XYZ'
person['age'] = 31
person['address'] = ['Jaipur']

print(person)
print(person['name'])
````

````python
# in and not in operations

print('name' in person)
print('fname' not in person)

#Length of the dictionary 
print(len(person))
````

````python
my_dict = {
    '1': 'data',
    '2': 'structure',
    '3': 'python',
    '4': 'programming',
    '5': 'language'
}

print(my_dict.get('1'))         # searches for a key and returns corresponding value
print(my_dict.items())          # returns a list of dictionary items
print(my_dict.keys())           # returns a list of dictionary keys
print(my_dict.values())         # returns a list of dictionary values          
````

````python
print(my_dict.pop('2'))         # removes key 2 and associated value
print(my_dict)

print(my_dict.popitem())        # removes the last key-value pair and returns it as a tuple
print(my_dict)

my_dict.clear()                 # removes all items 
print(my_dict)
````

````python
# Merging: merges one dictionary, d1 with another, d2 
# replaces corresponding key in d1 with value in d2,
# If key is missing in d1, new key-value pair from d2 is added to d1

d1 = {'a': 10, 'b': 20, 'c':30}
d2 = {'b': 200, 'd': 400}

d1.update(d2)
print(d1)
````

#### Sets

Unordered collection of hashable objects, iterable, mutable, contains unique elements (no duplicates)

````python
# a `set()` returns a set object from an iterable
x1 = set(['and', 'python', 'data', 'structure'])
print(x1)
print(type(x1))
````

````python
x = {'data', 'structure', 'and', 'python'}
print(len(x))               # length
print('structure' in x)     # membership
````

````python
x1 = {'data', 'structure'}
x2 = {'python', 'java', 'c', 'data'}

# Union
x3 = x1 | x2
print(x3)
print(x1.union(x2))

# Intersection
print(x1.intersection(x2))
print(x1 & x2)
````

````python
# difference
print(x1.symmetric_difference(x2))
print(x1 ^ x2)

# subset
print(x1.issubset(x2))
print(x1 <= x2)
````

##### Named tuples 
creates a tuple with named fields similar to regular tuples.

syntax:


````python
from collections import namedtuple

# Create Book namedtuple "class" 
Book = namedtuple('Book', ['name', 'ISBN', 'quantity'])   

# Create instance 
Book1 = Book(
    'Hands-On Data Structures and Algorithms with Python',  
    '9781788995573', '50')                                  

# Accessing data items
print('Using index ISBN:' + Book1[1])
print('Using key ISBN:' + Book1.ISBN)
````

#### Deques (double-ended queues)

Doubly-linked lists that provide efficient adding and removing of items
from both ends of the list.

````python
from collections import deque

s = deque()      # Creates empty deque
print(s)

my_queue = deque([1, 2, 'Name'])
print(my_queue)
````

````python
my_queue.append('age')        # Insert 'age' at the right end of the list.
print(my_queue)

my_queue.appendleft('age')    # Insert 'age' at the left end of the list.
print(my_queue)

my_queue.pop()                # Delete the rightmost value.
print(my_queue)

my_queue.popleft()            # Delete the leftmost value.
print(my_queue)
````

#### Ordered dictionaries
An ordered dictionary is a dictionary that preserves the order of the keys that are inserted. If the key order is important for any application, then `OrderedDict()` can be used:

```python
od = OrderedDict([items])
```

````python
from collections import OrderedDict

od = OrderedDict({'my': 2, 'name': 4, 'is': 2, 'Mohan': 5})
od['hello'] = 4
print(od)           # output dict in the same order and appends ('hello', 4)
````

#### Default dictionary
A dictionary subclass that returns default values for missing keys

````python
from collections import defaultdict

dd = defaultdict(int)
words = str.split('data python data data structure data python')
for word in words:
    dd[word] += 1
print(dd)
````

````python
# using te dict() function returns an error
d = dict(int)
words = str.split('data python data data structure data python')
for word in words:
    d[word] += 1
print(d)
````

#### ChainMap object
A dictionary that merges multiple dictionaries.

````python
from collections import  ChainMap
dict1 = {'data': 1, 'structure': 2}
dict2 = {'python': 3, 'language': 4}
chain = ChainMap(dict1, dict2)

print(chain)
print(list(chain.keys()))
print(list(chain.values()))
print(chain['data'])
print(chain['language'])
````

#### Counter objects
A dictionary that returns the counts corresponding to their objects/key.

````python
from collections import Counter

inventory = Counter('hello')
print(inventory)
print(inventory['l'])
print(inventory['e'])
print(inventory['o'])
````

#### UserDict UserList Userstring

These data types are used to add more functionalities to their base data
structure, such as a dictionary, list, and string. And we can create
subclasses from them for custom dict/list/string.

````python
# UserList wraps list objects
from collections import UserList

# We cannot push to this user List
class MyList(UserList):
    def push(self, key):
        raise RuntimeError('Cannot insert in the list')

d = MyList([11, 12, 13])
d.push(2)
````

````python
# Strings can be considered as an array of characters
# A character is string of one length and acts as a container
# that wraps a string object.

# Create a custom append function for string

from collections import UserString
class MyString(UserString):
    def append(self, value):
        self.data += value
        
s1 = MyString('data')
print("Original:", s1)
s1.append('h')
````
